```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from utils.matplotlib_visualization_utils import plot_data1d_histogram, plot_densities
```

```python
arr_data = np.genfromtxt('data/test.csv')
```

```python
arr_x1 = np.genfromtxt('visual_output/marginal_density_x.csv')
arr_x2 = np.genfromtxt('visual_output/marginal_density_y.csv')
```

```python
proportions = [ 0.00527027038150278256, 0.01025836847255434590, 0.09760324684671536023, 0.38860905629706645394, 0.49825905800216105737 ]
#proportions = [ 2.90770726e-01, 2.05436121e-05, 5.87199400e-02, 3.58711745e-01, 2.91777046e-01 ]
```

```python
fig = plt.figure(figsize=(6*1.618,6))
ax = fig.add_subplot(1,1,1)
plot_densities(arr_x1, stacked=False, prop=proportions, ax=ax)
plot_data1d_histogram(arr_data[:,0], bins=100,normed=True, ax=ax);
```

```python
fig = plt.figure(figsize=(6*1.618,6))
ax = fig.add_subplot(1,1,1)
plot_densities(arr_x1, stacked=True, prop=proportions, ax=ax)
plot_data1d_histogram(arr_data[:,0], bins=100,normed=True, ax=ax);
```

```python
fig = plt.figure(figsize=(6*1.618,6))
ax = fig.add_subplot(1,1,1)
plot_densities(arr_x2, stacked=False, prop=proportions, ax=ax)
plot_data1d_histogram(arr_data[:,1], bins=100,normed=True, ax=ax);
ax.set_xlim(-1.0,2);
```

```python
fig = plt.figure(figsize=(6*1.618,6))
ax = fig.add_subplot(1,1,1)
plot_densities(arr_x2, stacked=True, prop=proportions, ax=ax)
plot_data1d_histogram(arr_data[:,1], bins=100,normed=True, ax=ax);
ax.set_xlim(-1.0,2);
```
