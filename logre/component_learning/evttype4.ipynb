```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from utils.matplotlib_visualization_utils import plot_data2d, plot_contour2d
```

```python
evttype=4

visual_output_root = '.'
scatter_fname = visual_output_root + '/evttype{0}_scatter.csv'.format(evttype)
kde2d_fname = visual_output_root + '/evttype{0}_kde.csv'.format(evttype)
akde2d_fname = visual_output_root + '/evttype{0}_akde.csv'.format(evttype)
gridsearch_fname = visual_output_root + '/evttype{0}_gridsearch.csv'.format(evttype)
```

```python
fig = plt.figure(figsize=(24,24))

ax = fig.add_subplot(2,2,1)
plot_data2d(scatter_fname, ax=ax, title=r'input sample', 
            marker='.', xlabel=r'$Z_1$', ylabel=r'$Z_2$')
ax = fig.add_subplot(2,2,3)
plot_contour2d(kde2d_fname, ax=ax, title=r'cross validated', 
               nlevels=10, colorbar=False, contour_fmt='%1.2f', 
               xlabel=r'$Z_1$', ylabel=r'$Z_2$')
ax = fig.add_subplot(2,2,4)
plot_contour2d(akde2d_fname, ax=ax, title=r'cross validated and adapted', 
               nlevels=10, colorbar=False, contour_fmt='%1.2f', 
               xlabel=r'$Z_1$', ylabel=r'$Z_2$')
```

```python
fig = plt.figure(figsize=(10,8))

ax = fig.add_subplot(1,1,1)
plot_contour2d(gridsearch_fname, ax=ax, title=r'least squares cross validation score', 
               nlevels=10, nlevels_f=200, colorbar=True, cmap=plt.cm.YlOrRd, contour_fmt='%1.2f', 
               xlabel=r'$h_1$', ylabel=r'$h_2$')
```
